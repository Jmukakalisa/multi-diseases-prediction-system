Importing the Dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from google.colab import drive
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# drive.mount('/content/drive')
# # loading the data to a data frame
# data_frame = pd.read_csv('/content/drive/My Drive/ai/breastcancer.csv')
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

1 --> Benign

0 --> Malignant

```python
data_frame.groupby('label').mean()
```

Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Support Vector Classifier

```python
svm_classifier = SVC(kernel='linear')
```

```python
#training the support vector Machine Classifier
svm_classifier.fit(X_train, Y_train)
```

model evaluation

classification report

```python
Y_prediction = svm_classifier.predict(X_test)
print(classification_report(Y_test, Y_prediction))
```

```python
# accuracy score on the training data
X_train_prediction = svm_classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
# accuracy score on the test data
X_test_prediction = svm_classifier.predict(X_test)
svm_test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
print('Accuracy score of the test data : ', svm_test_data_accuracy)
```

confusion matrix

```python
confMatrix = confusion_matrix(Y_test, Y_prediction)
print(confMatrix)
```

heatmap on svm

```python
sns.heatmap(pd.DataFrame(confMatrix), annot=True)
plt.show
```

Logistic Regression

```python
lregression_model = LogisticRegression()
```

```python
# training the Logistic Regression model using Training data

lregression_model.fit(X_train, Y_train)
```

Model Evaluation

Classification report

```python
Y_prediction = lregression_model.predict(X_test)
print(classification_report(Y_test, Y_prediction))
```

Accuracy Score

```python
# accuracy on training data
X_train_prediction = lregression_model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
# accuracy on test data
X_test_prediction = lregression_model.predict(X_test)
lr_test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
print('Accuracy on test data = ', lr_test_data_accuracy)
```

Confusion matrix

```python
confMatrix = confusion_matrix(Y_test, Y_prediction)
print(confMatrix)
```

heatmap

```python
sns.heatmap(pd.DataFrame(confMatrix), annot=True)
plt.show
```

models' comparison

```python
models = pd.DataFrame({
    'Model' : [ ' Support Vector Machine Classifier', 'Logistic Regression Classifier'],
    'Score' : [svm_test_data_accuracy, lr_test_data_accuracy]
})


models.sort_values(by = 'Score', ascending = True)
```

Building a Predictive System, we will consider using SVM classifier since it has the highest accuracy score

```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = svm_classifier.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The Breast cancer is Malignant')

else:
  print('The Breast Cancer is Benign')

```
